```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
data2=pd.read_csv("factor_of_happiness2018(2).csv")
data2.head()
```

```python
data1.sort_values(by=["Country"],inplace=True,ascending=True)
data1=data1.reset_index(drop=True)
data1=pd.DataFrame(data1[:156])
data=pd.concat([data1,data2],axis=1)
data.head()
```

```python
data.info()
```

```python
extracts=["Country","Happiness score","GDP per person, 2015-2017","Healthy life expectancy, 2015-2017","Social support, 2015-2017","Freedom to make life choices, 2015-2017","Generosity, 2015-2017, without adjustment for GDP per person","Perceptions of corruption, 2015-2017"]
data=pd.DataFrame(data[extracts])
data.head()
```

```python
data=data.rename(columns={"Healthy life expectancy, 2015-2017":"Life_expectancy","GDP per person, 2015-2017":"GDP","Social support, 2015-2017":"Social_support","Freedom to make life choices, 2015-2017":"Freedom_lifechoice","Generosity, 2015-2017, without adjustment for GDP per person":"Generosity","Perceptions of corruption, 2015-2017":"Corruption"})
data.info()
```

```python
data = data.dropna(how='any',axis=0)
data = data.reset_index(drop=True)
data
```

```python
data.info()
```

```python
happiness_score=pd.DataFrame(data["Happiness score"])
happiness_score.describe().round(2)
```

```python
f,axes=plt.subplots(3,1,figsize=(15,12))
sb.boxplot(happiness_score,orient="h",ax=axes[0])
sb.distplot(happiness_score,kde=True,hist=True,color="r",ax=axes[1])
sb.violinplot(happiness_score,ax=axes[2],color="g")
```

Let classify the happiness score (or life ladder) into 3 main categories:
  1. (less than 4) : Unhappy
  2. (betwwen 4 and 6):normal to likely happy
  3. (more than 6): really happy

```python
happiness=[]
for i in happiness_score["Happiness score"]:
    if i<4:
        happiness.append("Unhappy")
    elif 4<=i<=6:
        happiness.append("Normal")
    else:
        happiness.append("Happy")
pred_happiness= pd.DataFrame(happiness,columns=["Predicted happiness"])
pred_happiness=pd.DataFrame(pred_happiness["Predicted happiness"].astype('category'))
data=pd.concat([data,pred_happiness],axis=1)
data
```

```python
# Import essential models and functions from sklearn
from sklearn.model_selection import train_test_split

#using DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
#Using Naive Bayes Prediction
from sklearn.naive_bayes import GaussianNB
# Using Logistic Regression 
from sklearn.linear_model import LogisticRegression
# Using Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.tree import export_graphviz
import graphviz


# Recall the Legendary-Total Dataset
predictors=["GDP","Life_expectancy","Social_support","Freedom_lifechoice"]
x=pd.DataFrame(data[predictors])
y=pd.DataFrame(data["Predicted happiness"])

# Split the Legendary-Total Dataset into Train and Test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)

# Decision Tree using Train Data
dectree=DecisionTreeClassifier(max_depth=3)
dectree.fit(x_train,y_train)
#Create a Gaussian Classifier using Train Data
gaussian = GaussianNB()
gaussian.fit(x_train,y_train)
# Create Logistisc Regression model
logistic = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=123)
logistic.fit(x_train,y_train)
# Create Random Forest Classifier
forest = RandomForestClassifier(n_estimators=100, random_state=123)
forest.fit(x_train,y_train)

# Predict Happiness values using DecisionTree
y_train_pred_tree=dectree.predict(x_train)
y_test_pred_tree=dectree.predict(x_test)
# Predict Happiness values using Gaussian Classifier
y_train_pred_gauss=gaussian.predict(x_train)
y_test_pred_gauss=gaussian.predict(x_test)
# Predict Happiness using Logistic Regression
y_train_pred_logistic=logistic.predict(x_train)
y_test_pred_logistic=logistic.predict(x_test)
# Predict Happiness using Random Forest Classifier
y_train_pred_forest=forest.predict(x_train)
y_test_pred_forest=forest.predict(x_test)

# Check the Goodness of Fit (by TreeClssifier)
print("Goodness of fit by using DecisionTreeClassifier:")
print("\tThe prediction accuracy on Train set: ",dectree.score(x_train,y_train))
print("\tThe prediction accuracy on Test set: ",dectree.score(x_test,y_test))
print()
# Check the Goodness of Fit (by Gaussian Classifier)
print("Goodness of fit by using Gaussian Classifier:")
print("\tThe prediction accuracy on Train set: ",metrics.accuracy_score(y_train,y_train_pred_gauss))
print("\tThe prediction accuracy on Test set: ",metrics.accuracy_score(y_test,y_test_pred_gauss))
print()
# Check the Goodness of Fit (by Logistic Regression)
print("Goodness of fit by using Logistic Regression:")
print("\tThe prediction accuracy on Train set: ",metrics.accuracy_score(y_train,y_train_pred_logistic))
print("\tThe prediction accuracy on Test set: ",metrics.accuracy_score(y_test,y_test_pred_logistic))
print()
# Check the Goodness of Fit (by Random Forest Classifier)
print("Goodness of fit by using Random Forest Classifier:")
print("\tThe prediction accuracy on Train set: ",forest.score(x_train,y_train))
print("\tThe prediction accuracy on Test set: ",forest.score(x_test,y_test))
print()



```

Hence, overall prediction made by Random Forest Classifier is the best. Therefore we will use Random Forest Classifier as the prediction model.
